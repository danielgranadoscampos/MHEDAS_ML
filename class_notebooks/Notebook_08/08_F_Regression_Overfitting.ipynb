## OVERFITTING: EXAMPLES & EXERCISES

Below are different cases and proposals for exercises on the Overfitting phenomenon, very common in ML.

Overfitting is one of the most important issues in machine learning, if not the most important.

If our training data is too few, our machine will not be able to generalize the knowledge and will be incurring in underfitting. This is the case where we teach you only one breed of dog and we expect you to be able to recognize 10 other different breeds of dogs. The algorithm will not be able to give us a good result due to a lack of "raw material" to solidify its knowledge. It is also an example of "underfitting" when the machine recognizes everything it "sees" as a dog, be it a photo of a cat or a car.

On the contrary, if we rigorously train our machine with 10 breeds of dogs only brown in color and then show a photo of a white dog, our model will not be able to recognize it as a dog because it does not exactly meet the characteristics it learned (the color necessarily had to be brown). Here it is an overfitting problem.

Both the problem of fitting "under" and "over" the data are bad because they do not allow our machine to generalize the knowledge and they will not give us good predictions (or classification, or grouping, etc.)

## 1st exercise Overfitting and underfitting in regression

Polynomial regression is a nonlinear relationship between independent x and dependent y variables. Fitting such type of regression is essential when we analyze fluctuated data with some bends. In this post, we'll learn how to fit a curve with polynomial regression data and plot it in Python. We use Scikit-Learn, NumPy, and matplotlib libraries in this tutorial.

We'll start by loading the required modules for this tutorial.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

We need test data, and we can generate it as shown below. You can use your own dataset too.

```python
#simulate x with n=250 data
n = 250         # elements number
x = list(range(n))
x = [i/100 for i in x]

#define y as a sin() with a random fluctuations
def GetPolyData(x):
    return np.sin(x) + np.random.uniform(-.2, .2, n)

y = GetPolyData(x)

#convert x and y in arrays of Numpy
train_x = np.array(x)
train_y = np.array(y)


# We'll check x data visually by creating a scatter plot.

plt.scatter(train_x, train_y)
plt.show();

#this sample data can represent x = weeks after treatment, y =biomaker disease relapse
```

Next, we'll define the polynomial model with PolymonialFeatures class fit it on training data. But, we need to define a polynomial model with PolymonialFeatures class fit it on training data. What degree is necessary? Maybe a Simple regression Model (degree = 1)?

```python
# Next, we'll define the polynomial model with PolymonialFeatures class fit it on training data.

polyModel = PolynomialFeatures(degree = 1) #polynomial degree = 1
xpol = polyModel.fit_transform(train_x.reshape(-1, 1))
preg = polyModel.fit(xpol,train_y)


# We need a linear model and we'll define it and fit on training data. Then we'll predict the x data with the model.

liniearModel = LinearRegression(fit_intercept = True)
liniearModel.fit(xpol, train_y[:, np.newaxis])
polyfit = liniearModel.predict(preg.fit_transform(train_x.reshape(-1, 1)))

# Finally, we'll plot the fitted curve.

plt.scatter(train_x, train_y)
plt.plot(train_x, polyfit, color = 'red')
plt.show();
```

Is the model fit to the data? What could be done?

We can fit a polynomial of degree 10?

```python
# Next, we'll define the polynomial model with PolymonialFeatures class fit it on training data.

polyModel = PolynomialFeatures(degree = 10) #degree 10
xpol = polyModel.fit_transform(train_x.reshape(-1, 1))
preg = polyModel.fit(xpol,train_y)


# We need a linear model and we'll define it and fit on training data. Then we'll predict the x data with the model.

liniearModel = LinearRegression(fit_intercept = True)
liniearModel.fit(xpol, train_y[:, np.newaxis])
polyfit = liniearModel.predict(preg.fit_transform(train_x.reshape(-1, 1)))

# Finally, we'll plot the fitted curve.

plt.scatter(train_x, train_y)
plt.plot(train_x, polyfit, color = 'red')
plt.show();
```

```python
#Now with a degree 4

# Next, we'll define the polynomial model with PolymonialFeatures class fit it on training data.

polyModel = PolynomialFeatures(degree = 4)
xpol = polyModel.fit_transform(train_x.reshape(-1, 1))
preg = polyModel.fit(xpol,train_y)


# We need a linear model and we'll define it and fit on training data. Then we'll predict the x data with the model.

liniearModel = LinearRegression(fit_intercept = True)
liniearModel.fit(xpol, train_y[:, np.newaxis])
polyfit = liniearModel.predict(preg.fit_transform(train_x.reshape(-1, 1)))

# Finally, we'll plot the fitted curve.

plt.scatter(train_x, train_y)
plt.plot(train_x, polyfit, color = 'red')
plt.show();
```

## 2nd exercise (CLASSIFICATION)

Python script demonstrating overfitting in classification with an example:

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler

# Generate linear data with two classes
np.random.seed(seed=1)
X = np.random.randn(100, 2)
y = np.where(X[:, 0] > 0.5, 1, 0)
print(X.shape)
print(X)
print(y)


```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Standardize features for better performance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a model with high complexity (prone to overfitting)
model_overfit = LogisticRegression(C=1e10)  # Large regularization parameter

# Train the model
model_overfit.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_overfit = model_overfit.predict(X_test_scaled)

# Calculate confusion matrix for overfitting model
confusion_matrix_overfit = confusion_matrix(y_test, y_pred_overfit)
print("Confusion matrix for overfitting model:\n", confusion_matrix_overfit)



```

```python
# Create a model with lower complexity (less prone to overfitting)
model_balanced = LogisticRegression(C=1)  # Smaller regularization parameter

# Train the balanced model
model_balanced.fit(X_train_scaled, y_train)

# Predict on the test set with the balanced model
y_pred_balanced = model_balanced.predict(X_test_scaled)

# Calculate confusion matrix for balanced model
confusion_matrix_balanced = confusion_matrix(y_test, y_pred_balanced)
print("Confusion matrix for balanced model:\n", confusion_matrix_balanced)



```

```python
# Compare the diagonals of both confusion matrices
# Higher values on the diagonal indicate better accuracy
print("Accuracy comparison:")
print("Overfitting:", confusion_matrix_overfit.trace() / confusion_matrix_overfit.sum())
print("Balanced:", confusion_matrix_balanced.trace() / confusion_matrix_balanced.sum())

```

 Conclusion: The balanced model achieves better generalization (accuracy) on unseen data compared to the overfitting model.

## Building a small dataset

Let's create a sample of examples with two values x1 and x2, with two categories.
For category 0, the underlying probability distribution is a 2D Gaussian centered on (0,0), with width = 1 along both directions. For category 1, the Gaussian is centered on (1,1). We assign label 0 to category 0, and label 1 to category 1.

```python
def make_sample(nexamples, means=([0.,0.],[1.,1.]), sigma=1.):
    normal = np.random.multivariate_normal
    # squared width:
    s2 = sigma**2.
    # below, we provide the coordinates of the mean as
    # a first argument, and then the covariance matrix
    # which describes the width of the Gaussian along the
    # two directions.
    # we generate nexamples examples for each category
    sgx0 = normal(means[0], [[s2, 0.], [0.,s2]], nexamples)
    sgx1 = normal(means[1], [[s2, 0.], [0.,s2]], nexamples)
    # setting the labels for each category
    sgy0 = np.zeros((nexamples,))
    sgy1 = np.ones((nexamples,))
    sgx = np.concatenate([sgx0,sgx1])
    sgy = np.concatenate([sgy0,sgy1])
    return sgx, sgy
```

Here, we create a very small training sample with only 30 examples per category, and a test sample with 200 examples per category. We're using such a small training sample because, as will be shown in this post, small samples are very easy to overfit.

```python
sgx, sgy = make_sample(30) #training n=30
tgx, tgy = make_sample(200) #test n= 200
```

```python
# note how the two categories are plotted
# together in one go by providing the
# label array as color argument (c=sgy)
plt.scatter(sgx[:,0], sgx[:,1], alpha=0.5, c=sgy)
plt.xlabel('x1')
plt.ylabel('x2')
```

We see that with such a low number of examples, it is not obvious that the samples follow Gaussian probability density functions. Also, since the Gaussians are so close, it's going to be quite difficult to separate the two categories.

## Overfitting

Let's try anyway with a neural network from scikit-learn (see in https://scikit-learn.org/stable/modules/neural_networks_supervised.html ). Here is an explanation of the parameters I use below:

* three hidden layers with 50 neurons each. I've chosen this fairly complex configuration on purpose to illustrate overfitting, which occurs when the model is too complex for the amount of data in the training sample.
* relu activation, because relu makes the training easier in neural nets with hidden layers.
* an increased maximum number of iterations, so that the network has time to converge
* a fixed random seed so that you can get the exact same results as me, every time you run the code

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(50,50,50), activation='relu', max_iter=10000, random_state=1)
mlp.fit(sgx,sgy)
```

Now let's define a small function to plot our results. The function will plot the examples in the two categories, as well as the probability that an (x1,x2) point belongs to category 1 (black means that this probability is close to 1, and white to 0.)

```python
def plot_result(sample, targets, linrange=(-5,5,101)):
    xmin, xmax, npoints = linrange
    gridx1, gridx2 = np.meshgrid(np.linspace(xmin,xmax,npoints), np.linspace(xmin,xmax,npoints))
    grid = np.c_[gridx1.flatten(), gridx2.flatten()]
    probs = mlp.predict_proba(grid)
    plt.pcolor(gridx1, gridx2, probs[:,1].reshape(npoints,npoints), cmap='binary')
    plt.colorbar()
    plt.scatter(sample[:,0], sample[:,1], c=targets, cmap='plasma', alpha=0.5, marker='.')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
```

```python
plot_result(sgx,sgy) #classification with training
```

Very nice eagle, but the probability distribution is very far from being optimal. We see that at the frontier, the neural network does its best to follow the patterns of the training sample. It is able to do that because its large number of parameters make it very flexible and adaptive.  

But let's see what happens if we plot the probability distribution with the larger test sample:

```python
plot_result(tgx,tgy) #classification using the training set with the test set
```

There are many examples that are classified in the wrong category. This neural network is very good with the training sample, but it has lost its generality and is thus useless in practice.

**This is overfitting.**

What can be done to solve the overfitting problem in this classification?





## Fixing overfitting

Now let's try again, but with a much more simple network, with a single layer with five neurons. The network is trained with the small training sample, and displayed with the larger test sample:

```python
mlp = MLPClassifier(hidden_layer_sizes=(5,), activation='relu', max_iter=10000, random_state=1)
mlp.fit(sgx,sgy)
plot_result(tgx,tgy)
```

This time, overfitting is much less of an issue. The network does not have enough parameters to be able to follow the random patterns of the training sample. Therefore, it behaves quite well on the test sample.

Let's try something else. This time, we use the complex network, but we provide much more training data: 10,000 examples per category instead of 30.  

```python
sgx, sgy = make_sample(10000) #training
mlp = MLPClassifier(hidden_layer_sizes=(50,50,50), activation='relu', max_iter=10000, random_state=1)
mlp.fit(sgx,sgy)
plot_result(tgx, tgy)
```

The network took a bit longer to train but this time, there is enough training data to properly constrain the parameters of the network, and the classification performance is going to be good in general.

## So why do we need complex networks then?

Well, to describe complex data! And for these networks to be efficient in this task, we will need a lot of training data.

In this section, we'll build a complex dataset with a lot of data, and see how well we can classify it.

To build the dataset, we just reuse our make_sample function several times and concatenate the resulting samples:

```python
sgxa, sgya = make_sample(1000, ([0.,0],[3.,3.]), 0.3)
sgxb, sgyb = make_sample(1000, ([1.,1],[4.,4.]), 0.3)
sgxc, sgyc = make_sample(1000, ([5.,5.],[-2.,-2.]), 0.6)
sgxd, sgyd = make_sample(1000, ([-1,3.],[3.,-1.]), 0.3)

sgx = np.concatenate([sgxa,sgxb,sgxc,sgxd])
sgy = np.concatenate([sgya,sgyb,sgyc,sgyd])
```

```python
plt.scatter(sgx[:,0], sgx[:,1], alpha=0.5, c=sgy)
plt.xlabel('x1')
plt.ylabel('x2')
```

Now, let's build a small network and see if we can classify that.

```python
mlp = MLPClassifier(hidden_layer_sizes=(3,), activation='relu', max_iter=10000, random_state=1)
mlp.fit(sgx,sgy)
```

```python
plot_result(sgx,sgy,linrange=(-4,7,201))
```

The network does not have enough parameters to fit the training data.

**This is underfitting.**

Still, it does quite a good job with its three neurons.

Let's increase the number of neurons on the hidden layer a bit:

```python
mlp = MLPClassifier(hidden_layer_sizes=(5,), activation='relu', max_iter=10000, random_state=1)
mlp.fit(sgx,sgy)
plot_result(sgx,sgy,linrange=(-4,7,201))
```

Wow! 5 neurons is already enough to fit the data, but we're a bit lucky. With a different topology we could have missed a patch.

Let's increase the complexity of the model even further:

```python
mlp = MLPClassifier(hidden_layer_sizes=(50,50,50), activation='relu', max_iter=10000, random_state=1)
mlp.fit(sgx,sgy)
plot_result(sgx,sgy,linrange=(-4,7,201))
```

Still no overfitting. The network now shows a smooth boundary, and I'm pretty sure it would be able to adapt further if needed.
